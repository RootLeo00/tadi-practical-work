```python
import glob
import matplotlib.pyplot as plt

image_paths = glob.glob("results/perona/cameraman*.png")

# Calculate the number of rows and columns for the subplots
num_images = len(image_paths)
num_rows = 1 #int(num_images / 3) + (num_images % 3 > 0)
num_cols = num_images #min(num_images, 3)
# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))

# Iterate over the image paths and plot each image in a subplot
for col, image_path in enumerate(image_paths):
    image = plt.imread(image_path)
    axes[ col].imshow(image, cmap='gray')
    # axes[ col].set_title("")
    # axes[ col].set_xlabel("X")
    # axes[ col].set_ylabel("Y")

# Remove any empty subplots
for i in range(num_images, num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axes[row, col])

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
